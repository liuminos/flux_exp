```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
modelc = fits.open("H://codes/pyMilne/example_hinode_spatially_coupled/modelout_spatially_coupled.fits")[0].data
model = fits.open("H://codes/pyMilne/example_hinode_spatially_coupled/modelout_pixel-to-pixel.fits")[0].data
```

```python
model.shape
```

```python
Bz = model[:,:,0]*np.cos(model[:,:,1])
Bzc = modelc[:,:,0]*np.cos(modelc[:,:,1])
```

```python
plt.figure(figsize=[12,5])
plt.subplot(121)
plt.imshow(Bz.T,cmap='PuOr',origin='lower')
#plt.colorbar()
plt.subplot(122)
plt.imshow(Bzc.T,cmap='PuOr',origin='lower')
```

```python
plt.imshow(Bz.T-Bzc.T,cmap='PuOr',origin='lower')
```

```python
np.std(Bz)
```

```python
plt.figure(figsize=[12,5])
plt.subplot(121)
plt.imshow(model[:,:,1].T*180./np.pi,cmap='PuOr',origin='lower')
plt.title("Field inclination")
plt.colorbar()
plt.subplot(122)
plt.imshow(model[:,:,2].T*180./np.pi,cmap='PuOr',origin='lower')
plt.title("Field azimuth")
plt.colorbar()
```

```python
gamma = model[:,:,1]*180./np.pi
gamma = gamma.flatten()
gammac = modelc[:,:,1]*180./np.pi
gammac = gammac.flatten()
```

```python
plt.figure(figsize=[12,5])
plt.subplot(121)
plt.hist(gamma,density=True,bins=180)
plt.ylim([0,0.03])
plt.subplot(122)
plt.hist(gammac,density=True,bins=180)
plt.ylim([0,0.03])
plt.tight_layout()
```
